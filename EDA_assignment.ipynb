```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
features=pd.read_csv("features.csv")
stores=pd.read_csv("stores.csv")
sales=pd.read_csv("train.csv")
features
```

```python
sales.head()
```

```python
df=pd.merge(features,stores,on='Store')
```

```python
df=pd.merge(df,sales,on=['Store','Date','IsHoliday'])
df.head(10)
```

```python
df.tail(10)
```

```python
print("Shape:",df.shape)
print("Info:\n",df.info())
print("Describe:\n",df.describe())
```

```python
df.nunique()
```

```python
df.isnull().sum()
```

```python
df.MarkDown1.fillna(df.MarkDown1.mean(),inplace=True)
df.MarkDown2.fillna(df.MarkDown2.mean(),inplace=True)
df.MarkDown3.fillna(df.MarkDown3.mean(),inplace=True)
df.MarkDown4.fillna(df.MarkDown4.mean(),inplace=True)
df.MarkDown5.fillna(df.MarkDown5.mean(),inplace=True)
```

```python
df.isnull().sum()
```

```python
df.Date=pd.to_datetime(df.Date)
df.info()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df
```

```python
df.reset_index(drop=True,inplace=True)
df
```

```python
q1=df['Weekly_Sales'].quantile(0.25)
q3=df['Weekly_Sales'].quantile(0.75)
iqr=q3-q1
lb=q1-1.5*iqr
ub=q3+1.5*iqr
ol=df[(df['Weekly_Sales']<lb)|(df['Weekly_Sales']>ub)]
print("Number of outliers in Weekly_Sales:",len(ol))
```

```python
df['ws_capped']=df['Weekly_Sales'].clip(lb,ub)
sns.boxplot(df['ws_capped'])
```

```python
q1=df['Temperature'].quantile(0.25)
q3=df['Temperature'].quantile(0.75)
iqr=q3-q1
lb=q1-1.5*iqr
ub=q3+1.5*iqr
ol=df[(df['Temperature']<lb) | (df['Temperature']>ub)]
df['temperature_capped']=df['Temperature'].clip(lb,ub)
sns.boxplot(df['temperature_capped'])
```
